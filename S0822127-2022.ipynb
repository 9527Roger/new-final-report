```python
#x,y
def violin(col):
    fig = px.violin(rowdata, y=col, x="match", color="match", box=True, template = 'plotly_dark')
    return fig.show()
#x,y
def kde(col):
    grid = sns.FacetGrid(rowdata, hue="match", height = 6, aspect=2)
    grid.map(sns.kdeplot,col)
    grid.add_legend()
```

```python
clear_data1 = rowdata.drop(['date'],axis=1)
clear_data =clear_data1.drop(['match'],axis=1)
label = rowdata['match']
```

```python
clear_data
```

```python
x_features = clear_data
y_label = label
x_features
```

Bagging

```python
from sklearn.model_selection import train_test_split # 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score
X = clear_data
y = label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
dt = BaggingClassifier(DecisionTreeClassifier()) # 
dt.fit(X_train, y_train) # 
y_pred = dt.predict(X_test) # 

# 
feature_importances = dt.estimators_[0].feature_importances_
selected_feature_names = clear_data.columns.tolist()

# 
for feature_name, importance in zip(selected_feature_names, feature_importances):
    print(": {}, : {:.4f}".format(feature_name, importance))

# 
print(": {:.2f}%".format(dt.score(X_test, y_test) * 100))
print("F1 score: {:.2f}%".format(f1_score(y_test, y_pred) * 100))

print('----------------------------------------------------------')
bdt = BaggingClassifier(DecisionTreeClassifier()) # Bagging
bdt.fit(X_train, y_train) # 
y_pred = bdt.predict(X_test) # 

# 
feature_importances = bdt.estimators_[0].feature_importances_

# 
for feature_name, importance in zip(selected_feature_names, feature_importances):
    print(": {}, : {:.4f}".format(feature_name, importance))

# 
print("Bagging: {:.2f}%".format(bdt.score(X_test, y_test) * 100))
print("BaggingF1 score: {:.2f}%".format(f1_score(y_test, y_pred) * 100))
```

Boosting

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostClassifier # AdaBoost
dt = DecisionTreeClassifier() # AdaBoost
ada = AdaBoostClassifier(dt) # AdaBoost
# 
ada_param_grid = {
"n_estimators" :[1,2,5,10],
"learning_rate": [0.0001, 0.001, 0.01, 0.1, 0.2, 0.3,1.5]}
ada_gs = GridSearchCV(ada,param_grid = ada_param_grid,scoring="f1", n_jobs= 10, verbose = 1)
ada_gs.fit(X_train,y_train) # 
ada_gs = ada_gs.best_estimator_ # 
y_pred = ada_gs.predict(X_test) # 
print("Adaboost: {:.2f}%".format(ada_gs.score(X_test, y_test)*100))
print("AdaboostF1 score: {:.2f}%".format(f1_score(y_test, y_pred)*100))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score

#  AdaBoost 
dt = DecisionTreeClassifier()  #  AdaBoost 
ada = AdaBoostClassifier(dt)  # AdaBoost 

# 
for feature_index, feature_name in enumerate(clear_data.columns[0:]):
    X_single_feature = clear_data[[feature_name]].values  # 
    ada.fit(X_single_feature, y)  #  AdaBoost 
    y_pred = ada.predict(X_single_feature)  # 

    # 
    print(": ", feature_name)
    print("Adaboost : {:.2f}%".format(ada.score(X_single_feature, y) * 100))
    print("Adaboost  F1 score: {:.2f}%".format(f1_score(y, y_pred) * 100))
    print()
```
