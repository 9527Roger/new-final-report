```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px
```

```python
num_cols = [col for col in rowdata.columns if rowdata[col].dtypes !='object']
```

```python
plt.figure(figsize =(20,10))
plot_number = 1
for column in num_cols:
    if plot_number <= 10:
        ax = plt.subplot(3, 5, plot_number)
        sns.distplot(rowdata[column])
        plt.xlabel(column)

    plot_number += 1
plt.tight_layout()
plt.show()
```

```python
#繪製小提琴圖形，以發生可能為x軸,y軸為想了解球員打擊表現跟球隊勝利發生相關的數據
def violin(col):
    fig = px.violin(rowdata, y=col, x="match", color="match", box=True, template = 'plotly_dark')
    return fig.show()
#繪製兩條曲線圖，以x軸為想了解球員打擊表現跟球隊勝利相關的數據,y軸為發生密度，兩條曲線一條為發生的曲線，另一條為不發生的曲線
def kde(col):
    grid = sns.FacetGrid(rowdata, hue="match", height = 6, aspect=2)
    grid.map(sns.kdeplot,col)
    grid.add_legend()
```

```python
clear_data1 = rowdata.drop(['date'],axis=1)
clear_data =clear_data1.drop(['match'],axis=1)
label = rowdata['match']
```

Bagging去討論選手與勝負的重要性

```python
from sklearn.model_selection import train_test_split # 拆分資料集
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score
X = clear_data
y = label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
dt = BaggingClassifier(DecisionTreeClassifier()) # 只使用一棵決策樹
dt.fit(X_train, y_train) # 擬合模型
y_pred = dt.predict(X_test) # 進行預測

# 取得第一個決策樹的特徵重要性
feature_importances = dt.estimators_[0].feature_importances_
selected_feature_names = clear_data.columns.tolist()

# 顯示每個特徵的重要性
for feature_name, importance in zip(selected_feature_names, feature_importances):
    print("特徵名稱: {}, 重要性: {:.4f}".format(feature_name, importance))

# 顯示結果
print("决策樹測試準確率: {:.2f}%".format(dt.score(X_test, y_test) * 100))
print("决策樹測試F1 score: {:.2f}%".format(f1_score(y_test, y_pred) * 100))

print('----------------------------------------------------------')
bdt = BaggingClassifier(DecisionTreeClassifier()) # 樹的Bagging
bdt.fit(X_train, y_train) # 擬合模型
y_pred = bdt.predict(X_test) # 進行預測

# 取得第一個決策樹的特徵重要性
feature_importances = bdt.estimators_[0].feature_importances_

# 顯示每個特徵的重要性
for feature_name, importance in zip(selected_feature_names, feature_importances):
    print("特徵名稱: {}, 重要性: {:.4f}".format(feature_name, importance))

# 顯示結果
print("决策樹Bagging測試準確率: {:.2f}%".format(bdt.score(X_test, y_test) * 100))
print("决策樹Bagging測試F1 score: {:.2f}%".format(f1_score(y_test, y_pred) * 100))
```

Boosting選手與勝負的關係

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostClassifier # 導入AdaBoost
dt = DecisionTreeClassifier() # 選擇決策樹作為AdaBoost的基礎模型
ada = AdaBoostClassifier(dt) # AdaBoost模型
# 使用網格搜索搜尋最佳化參數
ada_param_grid = {
"n_estimators" :[1,2,5,10],
"learning_rate": [0.0001, 0.001, 0.01, 0.1, 0.2, 0.3,1.5]}
ada_gs = GridSearchCV(ada,param_grid = ada_param_grid,scoring="f1", n_jobs= 10, verbose = 1)
ada_gs.fit(X_train,y_train) # 擬合模型
ada_gs = ada_gs.best_estimator_ # 最佳模型
y_pred = ada_gs.predict(X_test) # 預測
print("Adaboost測試準確率: {:.2f}%".format(ada_gs.score(X_test, y_test)*100))
print("Adaboost測試F1 score: {:.2f}%".format(f1_score(y_test, y_pred)*100))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score

# 建立 AdaBoost 模型
dt = DecisionTreeClassifier()  # 選擇決策樹作為 AdaBoost 的基礎模型
ada = AdaBoostClassifier(dt)  # AdaBoost 模型

# 迭代處理每個特徵
for feature_index, feature_name in enumerate(clear_data.columns[0:]):
    X_single_feature = clear_data[[feature_name]].values  # 提取單個特徵作為二維數組
    ada.fit(X_single_feature, y)  # 訓練 AdaBoost 模型
    y_pred = ada.predict(X_single_feature)  # 預測目標變量

    # 輸出結果
    print("特徵名稱: ", feature_name)
    print("Adaboost 訓練準確率: {:.2f}%".format(ada.score(X_single_feature, y) * 100))
    print("Adaboost 訓練 F1 score: {:.2f}%".format(f1_score(y, y_pred) * 100))
    print()
```
